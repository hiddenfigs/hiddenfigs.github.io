## Mark Down File Creator -- Hidden Figure Data Base Enteries
This Jupyter Notebook reads in a spread sheet of database enteries and then turns these enteries into seperate markdown files to be entered into the database. 

```python
#Packages to import
#Configures matplotlib to show figures in notebook
%matplotlib inline 
from pylab import *
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
import os
import statistics
```

```python
#Function that Reads in the gogole sheets file and makes a list that contains lists of the data base enteries
#file needs to be tab deliminated (.tsv)
def readIn(path):
    dataEnts = []
    #titles = []
    with open(path, encoding = "utf-8") as in_file:
       dataEnts =  stripSheetLine(in_file)

    return dataEnts

def stripSheetLine(in_file):
    lst = []
    for ent in in_file:
        
        entry = ent.strip().split("\t") #strips the new line out of the line and then splitting the enteries by commas

        lst.append(entry)
        
    return lst
    
```

```python
#list of the titles from the markdown sheet
titles = ["name", "linktitle", "last", "institution_of_phd","field_of_phd", "year_of_phd","researchAreas","relevantCourses","relevant_concepts", "wikipedia", "image", "general_bio", "key_contributions", "citations","layout", "level"] 
```

```python
# Define what the headers are for the mark down file -> it should be the same as the file
titles = data[0]
#remove titles from list to have the enteries just be in the data list
data.remove(titles)
#print(titles)
#print(data)
```

```python
#This function takes each entry makes a markdown file that is correctly configured to be added to the database 
def makeMDs(titles, entLst,imgPath):
    
    #Takes each entery in the entry list
    for ent in entLst:
        #Opens a new file to be written to... Remember ent is techincally a list of information
        print(ent)
        #The file will be name with the name of the entry
        entFile = open(ent[0] + ".md", "w")
        entFile.write("{\n")
        entFile.write("\"title\":\"name\",\n") #Steve told me to add this into here

        
        for title in titles:
            if title == "image": #Updated to reflect new syntax
               
                print("\"" + title + "\": [\"" + imgPath+ ent[titles.index(title)] + "\"],\n")
                entFile.write("\"" + title + "\":\"" + imgPath+ ent[titles.index(title)] + "\",\n")
            elif title == "researchAreas" or title == "relevantCourses" or title == "level":
                #These multiple enteries should be seperated by a / in order to know how to split them

                #Intialize a list that contains all enteries
                lst = ent[titles.index(title)].split("/")

                #the first half of the string to be written out to the file
                str = "\""+ title + "\":["
                for e in lst:
                    #We need to put a comma a the end of the entry if there is more to come but not if it is the last in the list
                    if lst.index(e) != len(lst) - 1:
                        str +=  "\""+ e + "\","
                    else:
                        str += "\"" + e + "\""

                str += "],\n"
                entFile.write(str)
            elif title == "citations":
            #seperate conditions are $ for citations
                #Intialize a list that contains all enteries
                lst = ent[titles.index(title)].split("$")

                #the first half of the string to be written out to the file
                str = "\""+ title + "\":["
                for e in lst:
                    if lst.index(e) != len(lst) -1: #If not in the last in the list
                        str +=  "\""+ e + "\","
                    else:
                        str += "\"" + e + "\""
                

                str += "],\n"
                entFile.write(str)

            elif title == "relevant_concepts":
                entFile.write("\n")
            #Formattings if there are more than one key contribution.. these are semi-formatted in the actual tab-delimited file    
            elif title == "key_contributions":
                lst = ent[titles.index(title)].split('/')

                str = "\"" + title +"\":{\n"

                for e in lst:
                    if lst.index(e) != len(lst)-1: #If not last in the list
                        str += e + ",\n"
                    else:
                        str += e + "},\n"  #If last in list
                entFile.write(str)
    
            

            
            else: #Normal entry format.. No listing or paths
                entFile.write("\"" + title + "\":\"" + ent[titles.index(title)] + "\",\n")
        entFile.write("\n}")
        entFile.close()


#I believe this accounts for everything here and follows standard formatting

        
```

```python
imgPathRepos = "/img/uploads/"

makeMDs(titles, data, imgPathRepos)
```
